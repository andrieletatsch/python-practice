**Exercise 5.1** — Countdown.

Print a countdown from 10 to 1, then print `'Go!'`.

**Exercise 6.1** — Find the first negative number.

## 📌 Section 7 — Try / Except

**Exercise 7.1** — Safe division.

Ask the user for two numbers and divide the first by the second.
Handle the case where the user tries to divide by zero.

```python
# Use try/except to handle ZeroDivisionError
# If division by zero, print a friendly message

a = int(input('Enter the first number: '))
b = int(input('Enter the second number: '))

try:
    result = a / b
    print(f'Result: {result}')
except ZeroDivisionError:
    print('Error: cannot divide by zero!')
```

**Exercise 7.2** — Safe list access.

Try to access an index that may not exist in the list. Handle the `IndexError`.

```python
# Use try/except to handle IndexError
# If the index does not exist, print 'Index out of range!'

fruits = ['apple', 'banana', 'cherry']

index = int(input('Enter an index to access: '))

try:
    print(f'Fruit: {fruits[index]}')
except IndexError:
    print('Index out of range!')
```

## 🔀 Section 8 — Mix It Up!

**Exercise 8.1** — Filter and classify sales.

Given a list of sales values:
- Skip any sale equal to 0 (use `continue`)
- Classify each valid sale as `'Low'` (< 100), `'Medium'` (100–299), or `'High'` (>= 300)
- Print each sale with its classification

```python
# Loop through sales
# Skip zeros with continue
# Classify and print each valid sale

sales = [50, 0, 320, 150, 0, 85, 400, 200, 0, 75]

#- Skip any sale equal to 0 (use `continue`)
for sale in sales:
    if sale  == 0:
        continue
    if sale < 100:
        category = 'Low'
    elif sale < 300:
        category = 'Medium'
    else:
        category = 'High'
    print(f'Sale: {sale} - {category}')

```

**Exercise 8.2** — Find the first product over budget.

You have a list of products with their prices. Loop through and find the first product that costs more than the budget. Print its name and price, then stop searching.

```python
budget = 100

products = [
    {'name': 'mouse', 'price': 45},
    {'name': 'keyboard', 'price': 89},
    {'name': 'monitor', 'price': 350},
    {'name': 'headset', 'price': 120},
]

# Loop through products
# Find the first product with price > budget
# Print its name and price, then break

for high_price in products:
    if high_price['price'] > budget:
        print(high_price['name'], high_price['price'])
        break
```

**Exercise 8.3** — Safe grade calculator.

The code below asks the user for 3 grades and handles invalid input.
Complete it: calculate the average and use an inline if to print if the student `Passed` (average >= 7) or `Failed`.

```python
grades = []

for i in range(1, 4):
    try:
        grade = float(input(f'Enter grade {i}: '))
        grades.append(grade)
    except ValueError:
        print('Invalid value. Using 0 for this grade.')
        grades.append(0)
total = 0
for g in grades:
    total += g
average = total / len(grades)
status = 'Passed' if average >= 7 else 'Failed'
print(f'Average: {average:.1f} — {status}')
```
