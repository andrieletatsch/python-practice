## **IF Else**

```python
#if

nota = 9
if nota >= 7.5:
    print("Parabens, voce foi aprovado!")
else:
    print("Infelizmente voce foi reprovado!")
```

```python
#if sem else

nota = 7.5
if nota >= 9:
    print("Parabens, voce foi aprovado!") #nao vai imprimir nada, pois a condicao nao e verdadeira


notaa = 9
if notaa >= 9:
    print("Parabens, voce foi aprovado!") #vai imprimir , pois a condicao e verdadeira
```

```python
#elif - else if - se a primeira condicao nao for verdadeira, entao testa a segunda condicao

nota = 5
if nota >= 9:
    print("Voce teve uma nota A!")
elif nota >= 7.5:
    print("Voce teve uma nota B!")
elif nota >= 6:
    print("Voce teve uma nota C!")   
else:
    print("Infelizmente voce foi reprovado!")
```

# **While Loop**

```python
#contagem regressiva

contador = 5
while contador > 0:
    print(contador)
    contador -=1
```

```python
#soma ate certo valor 

soma = 0
limite  = 10

while soma < limite:
    print(soma)
    numero = int(input('Digite um numero para somar: '))
    soma += numero  
    print(f"A soma : {soma}")
```

```python
#Adivinhe o numero 

numero_secreto = 42
tentativa = 0

while True:
   palpite = int(input("Adivinhe o numero secreto entre 1 e 100: "))
   tentativa += 1
   if palpite == numero_secreto:
       print(f"Parabéns! Você acertou o número secreto em {tentativa} tentativas.")
       break
   elif palpite < numero_secreto:
       print("Tente um número maior.")
   else:
       print("Tente um número menor.")
```

# **Break**

```python
#Break - e utilizado para interromper a execucao de um loop, seja ele um loop for ou while.
# Quando o comando break e encontrado dentro de um loop, a execucao do loop e imediatamente interrompida,
#  e o controle do programa passa para a linha de codigo imediatamente apos o loop.
cestaDeFrutas = ["maca", "Banana", "laranja", "Uva", "Abacaxi", "Melancia", "Pera", "Manga", "Goiaba", "Maracuja"]

for fruta in cestaDeFrutas:
    print(fruta)
    if fruta == "Banana":
        print("Banana encontrada! Parando a busca.")
        break
```

# **Continue**

```python
#continue - e utilizado para pular a iteracao atual do loop e passar para a proxima iteracao. 
# Ele permite que voce ignore certas partes do codigo dentro do loop com base em uma condicao especifica, sem sair completamente do loop.

cestaDeFrutas = ["maca", "Banana", "laranja", "Uva", "Abacaxi", "Melancia", "Pera", "Manga", "Goiaba", "Maracuja"]

for fruta in cestaDeFrutas:

   if fruta  == "Manga":
        print("Manga encontrada!Vamos pular para a proxima iteracao.")
        continue
   print(fruta)
```

# **Try Except** 

```python
#try except - e utilizado para lidar com excecoes, ou seja, erros que podem ocorrer durante a execucao de um programa.
# Ele permite que voce escreva codigo que pode gerar erros e forneca uma maneira de lidar com esses erros de forma controlada, evitando que o programa seja interrompido abruptamente.  

limite  = 10
acumulado = 0

while acumulado < limite:
    
    try:
        print(f"Acumulado: {acumulado}")
        numero = int(input("Informe um numero"))
        print(f"Numero: {numero}")
        acumulado += numero
    except ValueError:
        print("Valor invalido. Por favor, digite um numero inteiro.")

```
