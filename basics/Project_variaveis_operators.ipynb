## Objective: Learn how to use arithmetic operators in Python.

```python
#Arithmetic Operators
a=10
b=3

print("Addition:", a + b)
print("Subtration:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulo(remainder):", a % b)
print("Exponentiation:", a ** b)
```

```python
# Desafio: crie duas variáveis, calcule a média entre elas e imprima
x=5
y=8

print("Addition:", x + y)
print("Subtration:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)
print("Modulo(remainder):", x % y)
print("Exponentiation:", x ** y)

print("Average:", ((x + y)/2))
```
