Exercise 1

Crie:
- uma variável product_name
- uma variável price (float)
- uma variável quantity (int)

Calcule o valor total e imprima:
Use f-string.

```python
product_name = "Laptop"
price = 999.95
quantity = 1 

print(type(product_name))
print(type(price))
print(type(quantity))   

total = price * quantity 

print("Product:", product_name)

print("Unit Price: $"f"{price:.2f}")
#or
print(f"Unit Price: ${price:.2f}")

print("Quantity:", quantity)

print("Total: $"f"{total:.2f}")
#or
print(f"Total: ${total:.2f}")

```

# **Bloco 3 — Formatação Avançada**

**Exercise 6**

Crie:
usd_amount = 250
exchange_rate = 4.87

Converta para reais.

Mostre:

valor em dólar
taxa com 3 casas decimais
valor final com 2 casas

Use f-string.

```python
usd_amount = 250
exchange_rate = 4.87
brl_amount = usd_amount * exchange_rate

print(f"USD Amount: ${usd_amount:.3f}")
print(f"Exchange Rate: {exchange_rate:.3f}")
print(f"BRL Amount: R${brl_amount:.2f}")
```

**Exercise 7**

Refaça o exercício anterior usando:

% formatting

.format()

Você precisa praticar os três estilos

```python
usd_amount = 250
exchange_rate = 4.87
brl_amount = usd_amount * exchange_rate

# Using f-strings for formatted output
print(f"USD Amount: ${usd_amount:.3f}")
print(f"Exchange Rate: {exchange_rate:.3f}")
print(f"BRL Amount: R${brl_amount:.2f}")

# Alternative string formatting using the format() method
print("USD Amount: ${:0.3f}".format(usd_amount))
print("Exchange Rate: {:0.3f}".format(exchange_rate))
print("BRL Amount: R${:0.2f}".format(brl_amount))

# Using old-style string formatting - % formatting
print("USD Amount: $%.3f" %(usd_amount))
print("Exchange Rate: %f" %(exchange_rate))
print("BRL Amount: R$%.2f" %(brl_amount))
```

# **Bloco 5 — Desafio de Atenção (Erro Comum)** 


**Crie:**

- price = "100"
- discount = 20

Tente subtrair desconto do preço.

Você vai perceber que dá erro.

Corrija corretamente usando conversão.




```python
price = "100"
discount = 20 
price_int = int(price)

print(type(price))
print(type(price_int))

valor_desconto = price_int * (discount / 100)
print(valor_desconto)

final_price = price_int - valor_desconto 

print(f"Final price: ${final_price:.2f}")

# o erro que poderia ocerrer é o seguinte: TypeError: unsupported operand type(s) for *: 'str' and 'float', entao ja foi corrigido convertendo a string para inteiro usando int()
# da variavel price para price_int, e depois usando price_int para calcular o valor do desconto e o preço final.
```
