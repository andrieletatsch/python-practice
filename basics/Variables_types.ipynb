# **Compound Calculation**

Create:
 - base_salary
 - tax_percentage
 - bonus_percentage

Calculate:
- Tax amount
- Salary after tax
- Bonus amount (based on salary after tax)
- Final salary

Print everything formatted.

```python
base_salary = 6000.0
tax_percentage = 0.15
bonus_percentage = 0.25

tax_amount = base_salary * tax_percentage
salary_after_tax = base_salary - tax_amount
bonus_amount = salary_after_tax * bonus_percentage
final_salary = salary_after_tax + bonus_amount

print(f'Final salary: ${final_salary:.2f} (After a tax of ${tax_amount:.2f} and a bonus of ${bonus_amount:.2f})')
```

# **Multi-condition Logic**

Create:
- income
- credit_score
- years_employed

Loan rules:
- income >= 4000
- credit_score >= 700
- years_employed >= 2

But:
**If credit_score > 850, employment time does not matter.**

```python
income = 4000.0
credit_score = 851
years_employed = 1

is_eligible_for_loan = (income >= 4000 and credit_score >= 700 and (years_employed >= 2 or credit_score > 850))

print(f'Is eligible for loan: {"Yes" if is_eligible_for_loan else "No"}')
```
