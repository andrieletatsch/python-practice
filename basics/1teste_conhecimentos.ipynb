```python
##A lista [1, 2, 2, 3, 4, 4, 5] contém quantos elementos únicos?

#len conta a quantidade de elementos em um objeto, nesse caso, o conjunto de elementos únicos.

lista = [1, 2, 2, 3, 4, 4, 5]
elementos_unicos = set(lista)   
quantidade_unicos = len(elementos_unicos)

print(quantidade_unicos)    
```

```python
#Qual é o resultado da expressão for num in [1, 2, 3, 4, 5]: if num % 2 == 0: print(num)?

for num in [1,2,3,4,5]:
    if num % 2 == 0:
        print(num)

#O resultado da expressão acima é a impressão dos números pares da lista [1, 2, 3, 4, 5]. 
# O loop for percorre cada número na lista e verifica se o número é par usando a condição num % 2 == 0. Se a condição for verdadeira, o número é impresso. Portanto, os números 2 e 4 serão impressos como resultado.
```

```python
#Quantas vezes o loop for i in range (3 ): print(i) é executado?

for i in range(3):
    print(i)
#O loop for i in range(3): print(i) é executado 3 vezes.
# O range(3) gera uma sequência de números de 0 a 2, e o loop itera sobre essa sequência, imprimindo cada número.
#  Portanto, os números 0, 1 e 2 serão impressos como resultado.
```
