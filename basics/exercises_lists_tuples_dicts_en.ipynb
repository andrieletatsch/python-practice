**Exercise 3 — Slicing**

Use the list below:

```python
grades = [7.5, 8.0, 6.5, 9.0, 5.5, 8.5, 7.0, 6.0, 9.5, 8.0]
```

1. Print the first 3 grades
2. Print the last 3 grades using a negative index
3. Print grades from position 3 up to (not including) position 6
4. Print only the grades at even positions (0, 2, 4...) using step `[::2]`

````python
#1. Print the first 3 grades

grades = [7.5, 8.0, 6.5, 9.0, 5.5, 8.5, 7.0, 6.0, 9.5, 8.0]

grades[0:3]
print("First three grades:", grades[0:3])
````

````python
#2. Print the last 3 grades using a negative index
grades = [7.5, 8.0, 6.5, 9.0, 5.5, 8.5, 7.0, 6.0, 9.5, 8.0]
grades[-3:]

print("Last three grades:", grades[-3:])
````

````python
#3. Print grades from position 3 up to (not including) position 6
grades = [7.5, 8.0, 6.5, 9.0, 5.5, 8.5, 7.0, 6.0, 9.5, 8.0]
grades[3:6]

print("Grades from position 3 to 5:", grades[3:6])
````

````python
# 4. Print only the grades at even positions (0, 2, 4...) using step `[::2]`

grades = [7.5, 8.0, 6.5, 9.0, 5.5, 8.5, 7.0, 6.0, 9.5, 8.0]
grades[::2]

print("Grades at even positions:", grades[::2])
````

**Exercise 5 — Tuple vs List**

1. Create a tuple `coordinates = (-23.5505, -46.6333)` representing the latitude and longitude of São Paulo
2. Access and print each value separately using an f-string:
   `"Latitude: -23.5505 | Longitude: -46.6333"`
3. Try to change the first value of the tuple and observe the error
4. Write a comment explaining why the error occurs

````python
# 1. Create a tuple `coordinates = (-23.5505, -46.6333)` representing the latitude and longitude of São Paulo, Brazil.
coordinates = (-23.5505, -46.6333)

print("Coordinates of São Paulo:", coordinates)

# 2. Print the latitude and longitude separately.
print(f"Latitude: {coordinates[0]} | Longitude: {coordinates[1]}")

# 3. Try to change the first value of the tuple and observe the error
# Uncommenting the line below will raise a TypeError: 'tuple' object does not support item assignment
# coordinates[0] = -23.0000
print("Cannot modify tuple - tuples are immutable!")

# 4. Comment explaining why the error occurs:
# Tuples are immutable, meaning their values cannot be changed after creation.
# When we try to assign a new value to a tuple element, Python raises a TypeError.
# This is different from lists, which are mutable and allow item reassignment.
# The immutability of tuples makes them useful for data that should not be modified,
# and they can also be used as dictionary keys (unlike lists)
````

# **Block 3 — Dictionaries**

**Exercise 6 — Creating and accessing dictionaries**

Create a dictionary called `product` with the following information:
- `name`: `"Dell Notebook"`
- `price`: `3599.90`
- `stock`: `15`
- `available`: `True`

Then:
1. Print the product name
2. Print the price formatted to 2 decimal places using an f-string
3. Update the stock to `10`
4. Print the complete dictionary

````python
product = {
    "name": "Dell Notebook",
    "price": 3599.90,
    "stock": 15,
    "available": True
}
print(product) 

#1. Print the product name 
print(f"Product Name: {product['name']}")

#2. Print the product price to 2 decimal places using string formatting.
print(f"Product Price: ${product['price']:.2f}")

#3.Update the stock to 10
product["stock"] = 10
print(f"Updated stock: {product['stock']}")

#4.Print the complete dictionary:
print("Completed dictionary:", product)

````

**Exercise 7 — Adding and removing keys**

Use the dictionary below:

```python
student = {
    "name": "Andriele",
    "age": 25,
    "course": "Data Analysis"
}
```

1. Add the key `"final_grade"` with value `8.7`
2. Add the key `"approved"` with value `True` (based on grade >= 7)
3. Remove the key `"age"` using `del`
4. Print the final dictionary
5. Print all keys using `.keys()` and all values using `.values()`

````python
student = {
    "name": "Andriele",
    "age": 25,
    "course": "Data Analysis"
}

# 1. Add the key `"final_grade"` with value `8.7`
student['final_grade'] = 8.7
print("Updated student dictionary:", student)

#2.Add the key `"approved"` with value `True` (based on grade >= 7) 
student['approved'] = student['final_grade'] >=7
print("Student dictionary after adding approved Status:", student)

#3.Remove the key `"age"` from the dictionary using the `del` statement and print the updated dictionary.
del student['age']
print("Student dictionary after removing age:", student)

#4.Print the complete dictionary.
print("Final student dictionary:", student)

#5.Print all keys using `.keys()` and all values using `.values()`

# The .keys() method returns a view object that displays a list of all the keys in the dictionary. 
# The .values() method returns a view object that displays a list of all the values in the dictionary.  

print("Keys:", student.keys())
print("Values:", student.values())
````

**Exercise 10 — Movie Catalog**

Create a list of dictionaries called `movies` with at least 3 movies.

- `"title"`: string
- `"year"`: int
- `"rating"`: float (from 0 to 10)
