**Exercise 4.2** — Count items above average.

Given a list of temperatures, find how many are above the average.

```python
temperatures = [22, 30, 18, 25, 35, 20, 28]

# 1. Calculate the average (sum of temperatures divided by how many there are)
average = sum(temperatures) / len(temperatures)

# 2. Print the average
print(average)
# 3. Create a variable called count and set it to 0
count = 0

# 4. Use a for loop to go through each temperature
for temperature in temperatures:
    # 5. Inside the loop: if temperature is greater than the average, add 1 to count
     if temperature > average:
          count += 1 
# 6. After the loop, print how many temperatures were above average
print(f'Temperatures were above average: {count}')
```

## 📌 Section 5 — While Loop

**Exercise 5.1** — Save until goal.

You save $50 every week. Print how many weeks it takes to reach a savings goal.

```python
savings = 0
goal = 500
weekly_deposit = 50
weeks = 0

# 1. While savings is less than goal:
while savings < goal:
    # 2. Add weekly_deposit to savings
    savings += weekly_deposit
    # 3. Add 1 to weeks
    weeks += 1
    # 4. Print: 'Week 1: $50' (using f-string)
    print(f'Week:{weeks}: ${savings}')
# 5. After the loop, print: 'Goal reached in X weeks!'
print(f'Goal reached in {weeks} weeks!')
```

```python
savings = 0
goal = 500
weekly_deposit = 50
weeks = 0

# 1. While savings is less than goal:
while savings < goal:
    savings += weekly_deposit
    weeks += 1
    print(f'Week:{weeks} - ${savings}')
print(f'Goal reached in {weeks} weeks!')
```

**Exercise 5.2** — Password retry.

Keep asking for a password until the user gets it right (simulate with a variable).

```python
correct_password = 'python123'
attempts = 0
max_attempts = 3

# 1. While attempts is less than max_attempts:
while attempts < max_attempts:
    # 2. Ask the user to enter the password (use input())
    password = input('Please enter the password:')
    # 3. Add 1 to attempts
    attempts += 1
    # 4. If the entered password equals correct_password:
    if password == correct_password:
        # 5. Print 'Access granted!' and break
        print(f'Access granted! {attempts} attempts')
        break
    # 6. Else print: 'Wrong password. X attempt(s) remaining.'
    else:  
        print(f'Wrong password. {attempts} remaining.')

# 7. After the loop (outside while), if attempts equals max_attempts and no access:
if attempts == max_attempts and password != correct_password:
    # 8. Print 'Account locked. Too many failed attempts.'
    print(f'Account locked. Too many failed attempts')
```

## 🔀 Section 8 — Mix It Up!

**Exercise 8.1** — Classify and count temperatures.

Loop through temperatures, skip any below 0, classify the rest, and count how many are 'Hot'.

```python
temperatures = [22, -5, 35, 18, -2, 30, 12, 38, -1, 25]

hot_count = 0

# 1. Use a for loop to go through each temperature
for temperature in temperatures:
    # 2. If temperature is less than 0, use continue to skip it
    if temperature < 0:
        continue
    # 3. If temperature >= 30, category = 'Hot' and add 1 to hot_count
    if temperature >=30: 
        category = 'Hot'
        hot_count += 1
    # 4. Elif temperature >= 20, category = 'Warm'
    elif temperature >=20:
        category = 'Warm'
    # 5. Else category = 'Cool'
    else:
        category = 'Cool'
    # 6. Print: '22°C — Warm'
    print(f'Temperature: {temperature} — {category}')
# 7. After the loop, print: 'Total hot days: X'
print(f'Total hot days: {hot_count}')
```

**Exercise 8.2** — Shopping cart total with limit.

Add items to a cart total. Stop when the total exceeds the budget.

```python
budget = 150
cart_total = 0

items = [
    {'name': 'book', 'price': 25},
    {'name': 'pen set', 'price': 15},
    {'name': 'backpack', 'price': 80},
    {'name': 'notebook', 'price': 45},
    {'name': 'calculator', 'price': 60},
]

# 1. Use a for loop to go through each item
for item in items:
    # 2. Add the item price to cart_total
    cart_total += item['price']
    # 3. Print: 'Added: book ($25) | Total: $25'
    print(f"Added: {item['name']} (${item['price']}) | Total: ${cart_total}")
    # 4. If cart_total exceeds budget, print 'Budget exceeded! Stopping.' and break
    if cart_total > budget:
        print ('Budget exceeded! Stopping.')
        break
# 5. After the loop, print the final cart_total
print(f'Final total: ${cart_total}')
```

**Exercise 8.3** — Safe employee lookup.

Ask the user for an employee name, find them in the list, and print their info. Handle errors if something goes wrong.

```python
employees = [
    {'name': 'Alice', 'role': 'Data Analyst', 'salary': 5000},
    {'name': 'Bob', 'role': 'Engineer', 'salary': 6000},
    {'name': 'Carol', 'role': 'Manager', 'salary': 7500},
]

search = input('Enter employee name: ')
found = False

# 1. Use a for loop to go through each employee
for employee in employees:
    # 2. If employee name matches search (use ==):
    if employee['name'] == search:
        # 3. Use try:
        try:
            # 4. Print name, role and salary using f-string
            print(f"Name: {employee['name']} | Role: {employee['role']} | Salary: ${employee['salary']}")
            # 5. Set found = True and break
            found = True
            break
        # 6. Use except KeyError:
        except KeyError:
            # 7. Print 'Incomplete employee record!'
            print('Incomplete employee record!')
            # 8. Break
            break
# 9. After the loop, if found is False: print 'Employee not found.'
if found == False:
    print('Employee not found.')
```
